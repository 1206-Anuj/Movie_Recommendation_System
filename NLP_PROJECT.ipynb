# Recommendation System

A Recommendation System is a type of information filtering system that suggests relevant items to users based on their preferences, behavior, or past interactions.

It is widely used in platforms such as:
- E-commerce websites
- Streaming services
- Social media platforms
- Online learning platforms

The main goal of a recommendation system is to **predict user preferences and recommend items that the user is likely to be interested in**.

---

## Examples of Recommendation Systems

- Amazon recommending products based on past purchases
- Netflix recommending movies and shows
- Spotify recommending music playlists
- YouTube recommending videos

---

## Types of Recommendation Systems

### 1. Content-Based Filtering

Content-based filtering recommends items based on the **similarity between item features and user preferences**.

Example:
If a user likes **action movies**, the system will recommend other **action movies**.

Characteristics:
- Uses item features
- Based on user history
- No need for other users' data

---

### 2. Collaborative Filtering

Collaborative filtering recommends items based on **similar users or similar user behavior**.

Example:
If **User A and User B** like similar movies, then movies liked by **User B** may be recommended to **User A**.

Types of Collaborative Filtering:

- **User-Based Collaborative Filtering**
- **Item-Based Collaborative Filtering**

---

### 3. Hybrid Recommendation System

Hybrid systems combine **Content-Based Filtering and Collaborative Filtering** to improve recommendation accuracy.

Example:
Netflix uses a **hybrid recommendation system**.

---

# Preprocessing Steps

## Extracting Useful Information from Genres

The **genres** column in the dataset contains information about the categories of the movie such as Action, Comedy, Drama, etc.

However, the data in this column is stored in a **JSON-like format**, where each genre is represented as a dictionary containing an **id** and **name**.

Example:

[
{'id': 28, 'name': 'Action'},
{'id': 12, 'name': 'Adventure'},
{'id': 14, 'name': 'Fantasy'}
]

For building the recommendation system, we **do not need the id**.  
We only need the **genre names** because they describe the movie category.

### Goal

Extract only the **genre names** from the genres column.

Example Output:

['Action', 'Adventure', 'Fantasy']

These extracted genre names will later be **combined with other features (keywords, cast, overview, etc.)** to create a **tags column** used for calculating movie similarity.

````python
import ast
````

````python
data["genres"][0]
````

````python
def genre_extract(col):
    L=[]
    for i in ast.literal_eval(col):
        L.append(i["name"])
    return L
````

````python
data["genres"]= data["genres"].apply(genre_extract)
````

## Processing Cast Column

The **cast** column contains information about the actors who appeared in the movie.

The data is stored in a **JSON-like format**, where each entry contains details such as:
- cast_id
- character
- name
- gender

For building the recommendation system, we only need the **names of the actors**.

To keep the dataset simple and focused, we usually extract **only the top 3 main actors** from the cast list.

````python
data["cast"][0]
````

````python
import ast

def extract_cast(text):
    actors = []
    counter = 0
    
    for i in ast.literal_eval(text):
        if counter < 3:
            actors.append(i['name'])
            counter += 1
        else:
            break
            
    return actors

data['cast'] = data['cast'].apply(extract_cast)
````

````python
data["cast"][0]
````

## Processing Crew Column

The **crew** column contains information about the crew members involved in the movie, such as:

- Director
- Producer
- Writer
- Cinematographer
- Editor

For building a recommendation system, the **director is the most important crew member** because the director strongly influences the style and theme of the movie.

Therefore, we extract only the **director name** from the crew column.

````python
data["crew"][0]
````

````python
import ast

def extract_director(text):
    directors = []
    
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            directors.append(i['name'])
            break
            
    return directors

data['crew'] = data['crew'].apply(extract_director)
````

## Removing Spaces Between Names

In the Movie Recommendation System, we remove spaces between names so that multi-word names are treated as a **single token** during text vectorization.

If spaces are not removed, the model may treat **first name and last name as two different words**, which can affect the accuracy of similarity calculations.

### Example

Before removing spaces:

Sam Worthington → ['Sam', 'Worthington']

After removing spaces:

SamWorthington

Now the model treats it as **one unique feature**, which improves the quality of the recommendation system.

### Columns Where We Remove Spaces

We usually remove spaces from the following columns:

- genres
- keywords
- cast
- crew

````python
# Remove spaces between words in selected columns

data['genres'] = data['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
data['keywords'] = data['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
data['cast'] = data['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
data['crew'] = data['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
````



###  Text Vectorization (Bag of Words / CountVectorizer)


After completing preprocessing steps such as **stop word removal** and **stemming**, the next step is to convert text data into numerical form so that machine learning algorithms can understand it.

This process is called **Text Vectorization**.

One of the most commonly used techniques is **Bag of Words (BoW)**.

### What is Bag of Words?

Bag of Words converts text into a **vector of word frequencies**.

Steps:
1. Create a **vocabulary** of the most frequent words in the dataset.
2. Count how many times each word appears in each document.
3. Represent each document as a **numerical vector**.

### Example

Sentence 1:  
I love machine learning

Sentence 2:  
I love NLP

Vocabulary:

I | love | machine | learning | NLP

Vector Representation:

Sentence 1 → [1,1,1,1,0]  
Sentence 2 → [1,1,0,0,1]

In this project, we use **CountVectorizer from sklearn** to implement Bag of Words.

## Why We Use CountVectorizer to Implement Bag of Words

In Natural Language Processing (NLP), machine learning models cannot understand raw text directly.  
Therefore, we need to convert text into **numerical vectors**. One common technique for this is **Bag of Words (BoW)**.

### Why CountVectorizer?

`CountVectorizer` from **sklearn** is used to implement Bag of Words because it automatically performs several important tasks:

### 1️⃣ Vocabulary Creation
It scans the entire dataset and creates a **vocabulary of unique words**.

Example:

Text Data:
- "I love movies"
- "I love NLP"

Vocabulary created:
I | love | movies | NLP

---

### 2️⃣ Word Frequency Counting
It counts how many times each word appears in each document.

Example:

| Sentence | I | love | movies | NLP |
|--------|---|---|---|---|
| I love movies | 1 | 1 | 1 | 0 |
| I love NLP | 1 | 1 | 0 | 1 |

---

### 3️⃣ Converts Text into Numerical Matrix
It converts the text data into a **matrix of numbers** that machine learning algorithms can process.

Output shape example:

(4806 movies, 5000 words)

Each movie is represented as a **vector of word frequencies**.

---

### 4️⃣ Built-in Text Processing
`CountVectorizer` can also perform useful preprocessing automatically:

- Convert text to **lowercase**
- **Remove stop words**
- **Limit vocabulary size** using `max_features`
- Tokenize words automatically

## Cosine Similarity

After converting text data into numerical vectors using **Bag of Words**, we need a method to measure how similar two movies are.

For this purpose, we use **Cosine Similarity**.

### What is Cosine Similarity?

Cosine Similarity is a metric used to measure the **similarity between two vectors** by calculating the **cosine of the angle between them**.

Instead of comparing the magnitude of the vectors, it compares their **direction**.

### Cosine Similarity Formula

Cosine Similarity = (A · B) / (||A|| × ||B||)

Where:
- **A · B** → Dot product of vectors A and B
- **||A||** → Magnitude of vector A
- **||B||** → Magnitude of vector B

### Cosine Similarity Range

| Value | Meaning |
|------|------|
| 1 | Very similar |
| 0 | Not similar |
| -1 | Completely opposite |

In a **movie recommendation system**, we compute cosine similarity between movie vectors to find movies that are most similar to each other.

Movies with **higher similarity scores** are recommended to the user.

````python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
similarity.shape
````

## Movie Recommendation Function

After computing the **Cosine Similarity matrix**, the next step is to build a function that recommends movies based on similarity scores.

### How the Recommendation Function Works

1. Take a **movie title** as input.
2. Find the **index of the movie** in the dataset.
3. Retrieve the **similarity scores** for that movie.
4. Sort the movies based on **highest similarity score**.
5. Return the **top 5 most similar movies**.

Movies with the **highest cosine similarity scores** are considered the most similar and are recommended to the user.

````python
def recommend(movie):
    
    # find the index of the movie
    movie_index = new_df[new_df['title'] == movie].index[0]
    
    # get similarity scores for that movie
    distances = similarity[movie_index]
    
    # sort movies based on similarity score
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    # print recommended movie titles
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
````

````python
recommend("avatar")
````
